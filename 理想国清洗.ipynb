```python
with open('pg150.txt', 'r') as f:
    lines = f.readlines()
```

```python
len(lines)
```

```python
# str
str(lines)
```

```python
type(lines)
```

```python
# replace
p=str(lines).replace("\n", " ").replace("\r", " ").replace("*", " ").replace("!#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~", " ").replace("\'\\n\',", " ").replace("''", " ").replace('\\n\'', " ").split('/')
print(p)
```

```python
#strip
a=str(p).strip().replace("\\'", " ").replace("\'\'", " ").replace(",", " ").replace("space", " ").replace("\xef\xbb\xbf", " ").replace('""', ' ').replace("\'\w+", " ").replace("\s{2,}'", " ").lower().split()
print(a)
```

```python
from collections import Counter
b= Counter(a)
print(b)
```

```python
print(b.most_common(100))
```

```python
with open('结果.txt', 'w') as f:
    f.write(str(b.most_common(100)))
f.close()
```
